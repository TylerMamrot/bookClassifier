```python
import pandas as pd
import numpy as np
```

```python
## free file from https://simplemaps.com/data/us-cities.

df = pd.read_csv('../data/raw/uscities.csv')

df[:5]
    
```

```python
### basic visualizations
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

counts = df['state_id'].value_counts() 
#print(counts)
states = np.array([x for x in counts.axes])
c = [c for c in counts]



print(states)
print(c)


plt.figure(figsize=(120, 7)) 
ax.set_xticklabels(states[0], fontsize=7)
ax.bar(states[0], c, align='center')
```
